```python
import pandas as pd
```

```python
assembly = pd.read_csv('assembly_500_.csv')
defense = pd.read_csv('defense-diplomacy_500_.csv')
election = pd.read_csv('election_500_.csv')
north = pd.read_csv('north-korea_500_.csv')
politics = pd.read_csv('politics-general_500_.csv')
president = pd.read_csv('president_500_.csv')
```
