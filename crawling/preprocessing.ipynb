## 날짜를 string에서 datetime 타입으로 통일하기

### 경향

```python
kyung['date']
```

```python
from datetime import datetime, timedelta

datetime_string = "2023.01.22 14:08"
datetime_format = "%Y.%m.%d %H:%M"

datetime_result = datetime.strptime(datetime_string, datetime_format)
print(type(datetime_result)) # <class 'datetime.datetime'>
print(datetime_result) # 2021-12-31 13:35:42.657813

kyung['date'] = kyung['date'].apply(lambda _ : datetime.strptime(_,datetime_format))
kyung['date']
```

### 조선일보

```python
am_idx = cho[cho['date'].str.contains('오전')].index
pm_idx = cho[cho['date'].str.contains('오후')].index

datetime_string = "2023.01.24. 오전 10:43"
datetime_format_am = "%Y.%m.%d. 오전 %H:%M"
datetime_format_pm = "%Y.%m.%d. 오후 %H:%M"

cho.loc[am_idx,'date'] = cho.loc[am_idx,'date'].apply(lambda _ : datetime.strptime(_,datetime_format_am))
cho.loc[pm_idx,'date'] = cho.loc[pm_idx,'date'].apply(lambda _ : datetime.strptime(_,datetime_format_pm))
cho.loc[pm_idx,'date'] = cho.loc[pm_idx,'date'] + timedelta(hours=12)
```

```python
cho.head()
```
